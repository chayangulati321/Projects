Introduction : Counter Strike – Go (CSGO) dataset is inspired from a game. The dataset has total 17 columns and 1134 entries.

Attribute Information :

•	Map – On which area the team is playing like Dust I, Mirage, etc

•	Day – On which day the team is playing the game

•	Month – On which month the team is playing the game

•	Year – On which year the team is playing the game

•	Date – Proper date on which the game has been played

•	Wait Time – Waiting time of game, until all the players  has been joined to the respective teams

•	Match Time – Duration of the game played by the both teams

•	Team A Rounds – Total how many rounds did the team-A had played

•	Team B Rounds – Total how many rounds did the team-B had played

•	Ping – Time to response between teams and server

•	Kills – Total number of players killed during a match 

•	Assists – Total number of players helped other player to kill

•	Deaths – Total number of deaths count during the match

•	Mvp’s – Rating given at the end of match

•	HS% - Total number of head shots in a match

•	Points – Total number of points earned in a map

•	Result – Win or Lose or Tie is declared at the end of the match

```python
# Take Wait Time(s), Match Time(s), Team A Rounds, Team B Rounds, Ping, Kills, Assists, Deaths, Mvp's, HS%, Points attributes in X
X = df.iloc[:, 5:-2].values
np.random.seed(5)
```

# K-means clustering

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  #Inertia: It is the sum of squared distances of samples to their closest cluster center.
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter=300, n_init=10)
y_kmeans = kmeans.fit_predict(X)
print (y_kmeans)
```

```python
# Scatter plot to show the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
def Change(m):
    if m=='Win':
        return 0
    elif m=='Lost':
        return 1
    else:
        return 2
```

```python
df['Outcome'] = df['Result'].apply(Change)
```

```python
df.head()
```

```python
# In clustering generally we don't have labels but fortunately in this dataset we do have labels 
# So take advantage of this to know how our model perform
```

```python
# To check the performance of our model 
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(df['Outcome'], y_kmeans))
print()
print(classification_report(df['Outcome'], y_kmeans))
```

```python
# Metric for evaluating clustering algorithms
from sklearn.metrics import davies_bouldin_score

davies_bouldin_score(X, y_kmeans)
```

```python
# Metric for evaluating clustering algorithms
from sklearn.metrics import silhouette_score

print(silhouette_score(X, y_kmeans))
```

# Hierarchy Clustering

* Ward as linkage

```python
# Hierarchy Clustering with 'ward' as a method
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
print (y_hc)
```

```python
# Using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(8,6))

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

plt.title('Dendrogram')
plt.show()
```

```python
# Visualising the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of points')
plt.legend()
plt.show()
```

```python
# To check the performance of our model 
print(confusion_matrix(df['Outcome'], y_hc))
print()
print(classification_report(df['Outcome'], y_hc))
```

```python
# Metric for evaluating clustering algorithms
from sklearn.metrics import davies_bouldin_score

davies_bouldin_score(X, y_hc)
```

```python
# Metric for evaluating clustering algorithms
from sklearn.metrics import silhouette_score

print(silhouette_score(X, y_hc))
```
